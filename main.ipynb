```python
x=df[["Average Cost for two","Has Table booking","Has Online delivery","Price range"]]
#above 2 and below 2.5 poor
#above 2.5 and below 3.5 average
#above 3.5 and below 4 good
#above 4 and below 4.5 Very good
#above 4.5 and below 5 Excellent
```

```python
y=df["Aggregate rating"]
```

```python
x
```

```python
#encoding the data
from sklearn import preprocessing
labelencoder = preprocessing.LabelEncoder()
x.loc[:,"Has Table booking"] = labelencoder.fit_transform(x["Has Table booking"])
```

```python
x
#1 is yes,0 is no

x.loc[:, "Has Online delivery"] = labelencoder.fit_transform(x["Has Online delivery"])
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x=scaler.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,precision_score,recall_score,accuracy_score,confusion_matrix
def model_result(pred):
    print(f"MSE: {mean_squared_error(pred,y_test)}")
    print(f"MAE: {mean_absolute_error(pred,y_test)}")
    print(f"RMSE: {np.sqrt(mean_squared_error(pred,y_test))}")
    
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred_lr=model.predict(x_test)
y_pred_lr
```

```python
model_result(y_pred_lr)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
grid_params={'C':[0.001,0.01,0.1,1],'kernel':['rbf','poly','sigmoid'],'degree':[3,2,1]}
svr=GridSearchCV(SVR(),grid_params).fit(x_train,y_train)

```

```python
y_pred_svr=svr.predict(x_test)
model_result(y_pred_svr)
```

```python
from sklearn.tree import DecisionTreeRegressor
tree=DecisionTreeRegressor()
```

```python
param_grid={
    "max_depth":[10,30],
    "min_samples_split":[2,5],
    "min_samples_leaf":[1,2],
}
decisiontree=GridSearchCV(estimator=tree,param_grid=param_grid)
decisiontree.fit(x_train,y_train)
```

```python
y_pred_decisiontree=decisiontree.predict(x_test)
model_result(y_pred_decisiontree)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()
param_grid_rfr={'max_depth': [5,7,9,11], 'n_estimators': [10,15,20]}
```

```python
gridrfr=GridSearchCV(estimator=rfr,param_grid=param_grid_rfr)
gridrfr.fit(x_train,y_train)
```

```python
y_pred_rfr=gridrfr.predict(x_test)
model_result(y_pred_rfr)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor()
param_grid_knn={'n_neighbors':[15,17,19,20]}
grid_knn=GridSearchCV(knn,param_grid_knn)
grid_knn.fit(x_train,y_train)
y_pred_knn=grid_knn.predict(x_test)
```

```python
model_result(y_pred_knn)
```

```python
from sklearn.ensemble import AdaBoostRegressor
ada=AdaBoostRegressor()
param_grid_adaboost={'n_estimators':[50,100,200],"learning_rate":[0.1,0.5,1]}
adaboost=GridSearchCV(ada,param_grid_adaboost)
adaboost.fit(x_train,y_train)
ada_preds=adaboost.predict(x_test)
```

```python
model_result(ada_preds)
```
